# 6. Support Vector Machines - Linear Kernel

```python
conf_SVM_L, test_scoresSVM_L = classify(df.drop('y',axis=1), df['y'],classifiers['SVM'](kernel='linear'))
```

```python
conf_SVM_L_scaled, test_scores_SVM_L_scaled = classify(sonar_prepared_df, df['y'],classifiers['SVM'](kernel='linear'))
```

```python
param_grid = [
    {'C':2.**np.linspace(-2,15,18),'gamma':2.**np.linspace(-2,15,18)}
]

clf = SVC(kernel='linear')
grid_search_L = GridSearchCV(clf, param_grid, cv=10)
grid_search_L.fit(sonar_prepared_df, df['y'])
```

```python
conf_SVM_L_opt, test_scores_SVM_L_opt = classify(sonar_prepared_df, df['y'],
                                                 classifiers['SVM'](kernel='linear', C=grid_search_L.best_params_['C'],
                                                                        gamma = grid_search_L.best_params_['gamma']))
```

```python
cvres = grid_search_L.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(mean_score, params)
```

# 7. Support Vector Machines - RBF Kernel

```python
conf_SVM_RBF, test_scores_SVM_RBF = classify(df.drop('y',axis=1), df['y'],classifiers['SVM'](kernel='rbf', gamma='scale'))
```

```python
conf_SVM_RBF_scaled, test_scores_SVM_RBF_scaled = classify(sonar_prepared_df, df['y'],classifiers['SVM'](kernel='rbf', gamma='scale'))
```

```python
param_grid = [
    {'C':2.**np.linspace(-2,15,18),'gamma':2.**np.linspace(-2,15,18)}
]

clf = SVC(kernel='rbf')
grid_search_RBF = GridSearchCV(clf, param_grid, cv=10)
grid_search.fit(sonar_prepared_df, df['y'])
```

```python
conf_SVM_RBF_opt, test_scores_SVM_RBF_opt = classify(sonar_prepared_df, df['y'],
                                                     classifiers['SVM'](kernel='rbf', C=grid_search.best_params_['C'],
                                                                        gamma = grid_search.best_params_['gamma']))
```
