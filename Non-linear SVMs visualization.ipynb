# Random Generation for RBF Explanation
---

```python
# generate random points in a circle

import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


num_samples = 500

# make a simple unit circle 
theta = np.linspace(0, 2*np.pi, num_samples)
a, b = 1.1 * np.cos(theta), 1.1 * np.sin(theta)

# generate the points
# theta = np.random.rand((num_samples)) * (2 * np.pi)
r = np.random.rand((num_samples))
x, y = r * np.cos(theta), r * np.sin(theta)

ro = 2*np.random.rand((num_samples))
ro = ro[ro>1.2]
thetao = np.linspace(0, 2*np.pi, len(ro))

xo, yo = ro * np.cos(thetao), ro * np.sin(thetao)


z = x**2+y**2
zo = xo**2 + yo**2

# plots
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1,2,1)
ax.scatter(x, y,  label='Class -1')
ax.scatter(xo, yo,  label='Class 1')
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
# ax.plot(a, b, linestyle='-', linewidth=2, label='Hyperplane')

plt.ylim([-3,3])
plt.xlim([-3,3])
plt.grid(False)
plt.legend(loc='upper right')

ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.scatter(x,y,z, label='Class -1')
ax2.scatter(xo,yo,zo, label='Class 1')
plt.legend(loc='upper right')
ax2.grid(False)
ax2.w_xaxis.set_pane_color((0,0, 0, 0))
ax2.w_yaxis.set_pane_color((0,0, 0, 0))
ax2.w_zaxis.set_pane_color((0,0, 0, 0))

ax2.view_init(azim=50,elev=10)

ax2.set_xlabel('$X_1$')
ax2.set_ylabel('$X_2$')
ax2.zaxis.set_rotate_label(False)  # disable automatic rotation
ax2.set_zlabel('$X_1^2 + X_2^2$', rotation='horizontal',)

ax2.xaxis.set_ticks(np.arange(-2, 3, 1))
ax2.yaxis.set_ticks(np.arange(-2, 3, 1))
ax2.zaxis.set_ticks(np.arange(0, 5, 1))

    
plt.show()
```
